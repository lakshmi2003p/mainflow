```python
#scatterplot against income and expenditure
plt.scatter(data['income'],data['expenditure'])
#adding title to scatter plot
plt.title("Scatter Plot")
#adding X and Y labels
plt.xlabel('income')
plt.ylabel('expenditure')
plt.show()
```

```python
#Line Chart against income and expenditure
plt.plot(data['income'],data['expenditure'])
#adding title to scatter plot
plt.title("Line Chart")
#adding X and Y labels
plt.xlabel('income')
plt.ylabel('expenditure')
plt.show()
```

```python
#Bar Chart against income and expenditure
plt.bar(data['year'],data['own'])
#adding title to scatter plot
plt.title("Bar Chart")
#adding X and Y labels
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#Histogram
plt.hist(data['expenditure'])
plt.title("Histogram")
plt.show()
```

```python
#Pie chart
plt.pie(data['expenditure'])
plt.title("Pie Chart")
plt.show()
```
